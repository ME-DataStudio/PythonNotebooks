```python
#list comprehension
[(i,j) for i in range(10) for j in range(i+1,10) if not i%2]
```

```python
# Generator ipv list. Verschil is () ipv [].
g = ((i,j) for i in range(10) for j in range(i+1,10) if not i%2)
```

```python
g
```

```python
for i,j in g:
    print(i,j)
#Na het uitvoeren van deze statements is g leeg.
#Wil je het nogmaals gebruiken moet de generator nogmaals maken
#Voordeel van een generator is dat er weinig geheugenbeslag plaats vind in tegenstelling tot lists 
```

```python
for i in (j for j in range(100) if not j%2):
    print(i)
```

```python
#zip voegt dingen samen, bijvoorbeeld twee lijsten
#is ook een generator
alfabet='abcdefghijklmnopqrstuvwxyz'
nummers=range(1,27)
zipped=zip(alfabet,nummers)
```

```python
for i in zipped:
    print(i)
```

```python
#Na bovenstaande is zipped leeg, nogmaals uitvoeren geeft dan geen output
for i in zipped:
    print(i)
```
