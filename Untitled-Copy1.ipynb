## Functie definities

```python
def sliding_window(InArray):
    SizeInA = InArray.shape[0]
    OutArray=np.zeros(((SizeInA-3),3))
    #dan de lengte van iedere array in de data
    for j in range((SizeInA-2)):
        OutArray = np.vstack((InArray[j],InArray[j+1],InArray[j+2]))
    return OutArray
```

## Preprocessing
De dataset is zo niet bruikbaar om te leren. We gaan sliding window toepassen om 

```python
np.vstack((X_train[1856][0],X_train[1856][1],X_train[1856][2]))
```

```python
X_train_windows = np.array([sliding_window(array) for array in X_train])
```

```python
type(X_train_windows)
```

```python
X_train_windows[0:10]
```

### Omzetten naar clusters via KMeans
Eerst bepalen wat een goed aantal clusters is.

```python
max_K = 30

distortions =  np.zeros(max_K)
for K in range(max_K):
    kmeans = KMeans(n_clusters=K+1)
    kmeans.fit(np.vstack(X_train_windows))
    
    #distortions[K] = KMeans(n_clusters=K+1).fit(X).inertia_
    distortions[K] = kmeans.inertia_
```

```python
fig = plt.figure(figsize=(10,10), dpi = 100)
plt.plot(range(1,max_K+1), distortions)
plt.ylabel('Distortion')
plt.xlabel('Number of clusters')
plt.show()
```

We nemen zes clusters voor het clusteren van de windows. De knik zit bij drie, maar de distortion ligt nog vrij hoog. Bij zes blijkt ook nog een knikje te zitten.

```python
km = KMeans(n_clusters=15, init='k-means++', n_init=10, max_iter=300, tol=1e-04, random_state=0)
```

```python
labels = km.fit_predict(X_train_windows)
```
